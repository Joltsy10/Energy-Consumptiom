```python
import sys
sys.path.append('../src')
from evaluation import evaluate_model, print_metrics
import numpy as np
import torch
from model import LSTMModel
from data_preprocessing import load_data, prepare_data
from torch.utils.data import TensorDataset, DataLoader
```

```python
df = load_data('../data/household_power_consumption.txt')
(X_train, y_train), (X_val, y_val), (X_test, y_test), scaler = prepare_data(df)

X_train_tensor = torch.FloatTensor(X_train)
y_train_tensor = torch.FloatTensor(y_train)
X_val_tensor = torch.FloatTensor(X_val)
y_val_tensor = torch.FloatTensor(y_val)
X_test_tensor = torch.FloatTensor(X_test)
y_test_tensor = torch.FloatTensor(y_test)

train_dataset = TensorDataset(X_train_tensor, y_train_tensor)
val_dataset = TensorDataset(X_val_tensor, y_val_tensor)
train_loader = DataLoader(train_dataset, batch_size=64, shuffle=True)
val_loader = DataLoader(val_dataset, batch_size=64, shuffle=False)
```

```python
from sklearn.metrics import mean_absolute_error

print("\n" + "="*50)
print("BASELINE COMPARISON")
print("="*50)

y_test_kw = scaler.inverse_transform(y_test.reshape(-1, 1)).flatten()

naive_predictions_kw = y_test_kw[:-1]
actual_values_kw = y_test_kw[1:]

naive_mae = mean_absolute_error(actual_values_kw, naive_predictions_kw)

naive_mape = np.mean(np.abs((actual_values_kw - naive_predictions_kw) / actual_values_kw)) * 100

lstm_mae = 0.0804
lstm_mape = 9.76

print(f"\nNaive Baseline (predict last value):")
print(f"  MAE:  {naive_mae:.4f} kW")
print(f"  MAPE: {naive_mape:.2f}%")

print(f"\nLSTM Model:")
print(f"  MAE:  {lstm_mae:.4f} kW")
print(f"  MAPE: {lstm_mape:.2f}%")

if lstm_mae < naive_mae:
    improvement_mae = ((naive_mae - lstm_mae) / naive_mae * 100)
    print(f"\nImprovement:")
    print(f"  MAE:  {improvement_mae:.1f}% better than naive")
else:
    worse = ((lstm_mae - naive_mae) / naive_mae * 100)
    print(f"\nComparison:")
    print(f"  LSTM is {worse:.1f}% worse than naive (this shouldn't happen!)")

print("="*50 + "\n")
```
